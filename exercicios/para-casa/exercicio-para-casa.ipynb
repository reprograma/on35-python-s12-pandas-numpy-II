**Parte 3: Agrupamento e Agregação**

1. Agrupar os dados por categoria e calcular o total de vendas por categoria.
   
Categoria         Valor_Total
* 0    Alimentos     742457.74
* 1   Brinquedos    1187964.04
* 2  Eletrônicos    897261.71
* 3       Livros    963685.46

```python
total_por_categoria = df.groupby('Categoria')['Valor_Total'].sum().reset_index()
print(total_por_categoria)
```

2. Pergunta 6: Qual categoria teve o maior valor total de vendas?

**Categoria 'Brinquedos' com vendas totais de R$ 1187964.04**

```python
#Calcular o valor máximo por categoria
##Identificar a maior categoria em vendas
###Determinar (função max) o valor máximo em vendas por categorias
vendas_por_categoria = df.groupby('Categoria')['Valor_Total'].sum()
categoria_maior_venda = vendas_por_categoria.idxmax()
valor_maximo_venda = vendas_por_categoria.max()
print(f'A categoria com o maior valor de vendas é: {categoria_maior_venda}, com vendas totais de R$ {valor_maximo_venda:.2f}')
```

3. Calcular a quantidade média vendida por categoria

```python
#Média por categoria
media_quantidade_categoria = df.groupby('Categoria')['Quantidade_Vendida'].mean().reset_index()
print(media_quantidade_categoria)
```

4. Pergunta 7: Qual categoria teve a maior quantidade média vendida?
   
**A categoria com a maior quantidade média vendida é: Alimentos, com uma média de 55.62 unidades vendidas**

```python
vendas_por_categoria = df.groupby('Categoria')['Quantidade_Vendida'].mean()
categoria_maior_quantidade_media = vendas_por_categoria.idxmax()
quantidade_media_vendida = vendas_por_categoria.max()
print(f'A categoria com a maior quantidade média vendida é: {categoria_maior_quantidade_media}, com uma média de {quantidade_media_vendida:.2f} unidades vendidas.')
```

**Parte 4: Visualização de Dados**

1. Criar um gráfico de barras mostrando o total de vendas por categoria

```python
total_por_categoria = df.groupby('Categoria')['Valor_Total'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(total_por_categoria['Categoria'], total_por_categoria['Valor_Total'], color='skyblue')
plt.xlabel('Categoria')
plt.ylabel('Valor Total de Vendas (R$)')
plt.title('Total de Vendas por Categoria')
plt.show()
```

3. Criar um gráfico de linha mostrando a variação diária das vendas de 'Eletrônicos'

```python
vendas_diarias = eletronicos.groupby('Data')['Valor_Total'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(vendas_diarias['Data'], vendas_diarias['Valor_Total'], color='purple')
plt.xlabel('Data')
plt.ylabel('Valor Total de Vendas (R$)')
plt.title('Variação Diária das Vendas de Eletrônicos')
plt.xticks(ticks=vendas_diarias['Data'][::5], rotation=45)
plt.show()
```
