### 3. **Pandas: Consultas e Filtros com Datas**
**Exerc√≠cio**: Filtre os registros para mostrar apenas aqueles com `Data` entre '2023-05-01' e '2023-05-31'.


```python
df = df[(df['Data'] >= '2023-05-01') & (df['Data'] <= '2023-05-31')]
df.head(10)
```

```python
df = df.rename(columns={'Valor_1': 'estoque da loja1', 'Valor_2': 'Estoque da loja2', 'Valor_3': 'vendas'})
```
