```python
import pandas as pd

df = pd.read_csv('dados_ficticios.csv')
```

```python
df.describe()
```

```python
filtro = df[(df['Categoria'] == 'A') & (df['Valor_1'] > 60)]
filtro.head(20)
```

```python
df = df[(df['Data'] >= ('2023-05-01')) & (df['Data'] <= ('2023-05-31'))]
df.head(5)
```

```python
df = df.rename(columns={'Valor_1': 'estoque da loja1', 'Valor_2': 'Estoque da loja2', 'Valor_3': 'vendas'})
```

```python
df.head()
```

```python
agrupado_media = df.groupby('Categoria')[['estoque da loja1', 'Estoque da loja2']].mean()
agrupado_media
```

```python
agrupado_custom = df.groupby('Categoria').agg({
    'vendas': 'sum',
    'Categoria': 'count'
}).rename(columns={'Categoria': 'Contagem'})
```

```python
agrupado_custom
```
