```python
#df_data = pd.to_numeric(df['Data'])
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df = df[(df['Data'] >= '2023-05-01') & (df['Data'] <= '2023-05-31')]
df.head(10)
```

```python
df = df.rename(columns={'estoque da loja1':'Estoque da loja1', 'estoque da loja2': 'Estoque da loja2', 'vendas': 'Vendas'})
```

```python
df.head()
```

```python
agrupado_media = df.groupby('Categoria')[['Estoque da loja1', 'Estoque da loja2']].mean()

agrupado_media
```

```python
agrupado_max = df.groupby('Categoria')[['Estoque da loja1', 'Estoque da loja2']].mean()

agrupado_max
```

```python
df_agrupamento = df.groupby('Categoria').agg({
    'Vendas': 'sum',
    'Categoria': 'count'
}).rename(columns={'Categoria': 'Contagem'})
```

```python
df_agrupamento
```

```python
import matplotlib.pyplot as plt
```

```python
media_valor_1 = df.groupby('Categoria')['Estoque da loja1'].mean()
```

```python
media_valor_1.plot(kind='bar', color='#636', title='Média de Estoque da loja1 por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Média de Estoque da loja1')
plt.show()
```
